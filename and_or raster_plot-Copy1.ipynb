```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plotting 1 and 1 = 1
#firing rate for an input of 0 is 25spikes/sec, so we get 1/25 sec per spike
#firing rate for an input of 1 is 52spikes/sec, so we get 1/52 sec per spike
#firing rate for output is 54spikes/sec for 1 and 1, (retreived from predict function), 1/54
in1=np.arange(1/52,1,1/52)
in2=np.arange(1/52,1,1/52)
#add 0.01 sec to show delay
out=np.arange(1/54+0.01,1,1/54)

x = [in1,in2,out]

colors = np.array([[1, 0, 0],
                    [0, 1, 0],
                    [0, 0, 1]])
labels = ['input 1', 'input 2', 'output']
fig, axs = plt.subplots(1,1,figsize=(30,10))

# create a horizontal plot
axs.eventplot(x,color=colors)

axs.legend(labels, prop={'size': 25})
plt.show()
```
